```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("iris.txt", sep="\t")
iris.head()
```

```python
names = ("Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width")
species = iris["Species"].unique()
colors = ("red", "green", "blue")
i = 1
plt.figure(figsize=(20,20))
for col in names:
    for row in names:
        plt.subplot(4,4,i)
        if row == col:
            plt.hist(iris[col].values)
        else:
            for j, spec in enumerate(species):
                spec_cols = iris.loc[iris["Species"] == spec, [col]]
                spec_rows = iris.loc[iris["Species"] == spec, [row]]
                plt.scatter(spec_cols.values, spec_rows.values, color = colors[j])
        if (i-1) % 4 == 0:
            plt.ylabel(col)
        if i >12:
            plt.xlabel(row)
        i += 1
plt.show()
```
